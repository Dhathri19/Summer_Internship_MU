2.1 For transported feature

```python
# The number of different values in the "transported" column
transported_count = train['Transported'].value_counts()

transported_count
```

2.2 For age feature

```python
# Set the figure size
plt.figure(figsize=(10, 10))

# Filter the data for True and False values in the 'Transported' column
transported_true = train[train['Transported'] == True]
transported_false = train[train['Transported'] == False]

# Plot the histogram for True values
plt.hist(transported_true['Age'], bins=range(0, 50, 1), alpha=0.7, label='Transported: True')

# Plot the histogram for False values
plt.hist(transported_false['Age'], bins=range(0, 50, 1), alpha=0.7, label='Transported: False')

# Setting alpha = 0.7 --> the bars will have a slightly transparent appearance, allowing some visibility of the underlying elements or overlapping bars in the plot.

# Set the plot title and labels
plt.title('Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Count')

# Add a legend
plt.legend()

plt.show()
```

*   From this, we can say that in the age gap between, 0 - 18, are more likely to be transported.
*   From this, we can say that in the age gap between, 18 - 25, are less likely to be transported.
*   From this, we can say that in the age gap between, 25 - 50, are equally likely to be transported and not be transported.

For all expenditure features: RoomService, FoodCourt, Shopping mall, Spa and VR Desk and transported.

```python
# Expenditure features
exp_feats = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

# Create subplots
fig, axes = plt.subplots(5, 2, figsize=(10, 20))

# Iterate over the expenditure features
for i, var_name in enumerate(exp_feats):
    # Left plot
    ax = axes[i, 0]
    ax.hist(train[train['Transported'] == True][var_name], bins=30, alpha=0.5, label='Transported')
    ax.hist(train[train['Transported'] == False][var_name], bins=30, alpha=0.5, label='Not Transported')
    ax.set_title(var_name)
    ax.legend()

    # Right plot
    ax = axes[i, 1]
    ax.hist(train[train['Transported'] == True][var_name], bins=30, alpha=0.5, label='Transported')
    ax.hist(train[train['Transported'] == False][var_name], bins=30, alpha=0.5, label='Not Transported')
    ax.set_title(var_name)
    ax.set_ylim([0, 100])
    ax.legend()

# Adjust layout and spacing
fig.tight_layout()

# Show the plots
plt.show()
```



*   People who got transported, tend to spend less on these amenities.
*   There are a small number of outliers.

Create a feature that tracks expenditures across all the five features. Create a binary feature if a person has not spent anything.
Create a log feature to treat the outliers.



```python
# Categorical features
cat_feats = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']

# Create subplots
fig, axes = plt.subplots(len(cat_feats), 1, figsize=(10, 16))

# Iterate over the categorical features
for i, var_name in enumerate(cat_feats):
    # Plot count bars
    ax = axes[i]
    transported_counts = train[train['Transported'] == True][var_name].value_counts()
    not_transported_counts = train[train['Transported'] == False][var_name].value_counts()

    # Get unique categories
    categories = np.union1d(transported_counts.index, not_transported_counts.index)

    # Set the width of each bar
    bar_width = 0.35

    # Set the positions of the bars on the x-axis
    x_transported = np.arange(len(categories))
    x_not_transported = x_transported + bar_width

    # Plot the count bars
    ax.bar(x_transported, transported_counts.reindex(categories, fill_value=0), width=bar_width, label='Transported')
    ax.bar(x_not_transported, not_transported_counts.reindex(categories, fill_value=0), width=bar_width, label='Not Transported')

    # Set x-ticks and labels
    ax.set_xticks(x_transported + bar_width / 2)
    ax.set_xticklabels(categories)

    # Set title
    ax.set_title(var_name)

# Adjust layout and spacing
fig.tight_layout()

# Show the plots
plt.show()
```

*   VIP is not a useful feature since, since the distribution of true and false is not balanced.

We might consider dropping the VIP column to prevent overfitting.
